```python
from skimage import io
import numpy as np
from collections import defaultdict
import pandas as pd

def get_target_color_number(X):
    dic = defaultdict(int)
    for row in range(X.shape[0]):
        for col in range(X.shape[1]):            
            dic[tuple(X[row,col])] += 1
    return dic

import os
from PIL import Image

image_path = './seg/'
def read_img(image_path):
    l = []
    l2 = []
    for image_name in os.listdir(image_path):
        if image_name.endswith(".png"):
            print(image_path+image_name)
            #original_im = Image.open(image_path+image_name)
            img = io.imread(image_path+image_name)
            l.append(get_target_color_number(img))
            l2.append(image_name.split('_seg_1024.png')[0][7:])
    return l,l2
```
